##20.	Lectura y escritura de archivos
import os : Este comando importa el módulo os, que permite interactuar con el sistema operativo, incluyendo la manipulación de archivos y rutas.

def convertir_a_mayusculas():  Aquí se define una función llamada convertir_a_mayusculas. Esta función no recibe parámetros y se encargará de convertir el contenido de un archivo de texto a mayúsculas.

archivo_entrada: Esta variable contiene la ruta completa del archivo de texto que se desea leer. El prefijo r indica que es una cadena cruda, lo que permite que las barras invertidas se interpreten literalmente, evitando problemas con caracteres de escape.

archivo_salida: Esta variable construye la ruta para el archivo de salida.
os.path.dirname(archivo_entrada) obtiene el directorio del archivo de entrada.
os.path.join(...) combina este directorio con el nombre del nuevo archivo (salida.txt), asegurando que la ruta sea válida según el sistema operativo.
python


try:: Comienza un bloque de código donde se intentará ejecutar el código que sigue. Si ocurre un error, se saltará a la sección except.

with open(...) as infile:: Abre el archivo de entrada en modo lectura ('r'), especificando que se usará la codificación utf-8 para leer correctamente los caracteres.
infile: Este es un manejador de archivo que permite acceder al contenido del archivo. Usar with asegura que el archivo se cierre automáticamente después de su uso.
python


contenido: Esta variable almacena todo el texto leído del archivo de entrada. El método read() obtiene el contenido completo como una única cadena.

contenido_mayusculas: Esta variable almacena el contenido convertido a mayúsculas. Se utiliza el método upper() para transformar todas las letras de la cadena contenido en mayúsculas.
python

with open(...) as outfile:: Abre (o crea) el archivo de salida en modo escritura ('w'). Esto significa que si el archivo ya existe, se sobrescribirá.
outfile: Este manejador permite escribir en el archivo de salida.


outfile.write(...): Escribe el contenido en mayúsculas en el archivo de salida. Aquí, contenido_mayusculas se guarda en salida.txt.


print(...): Muestra un mensaje en la consola confirmando que el proceso de conversión se realizó correctamente y que el contenido se ha escrito en el archivo de salida.


except FileNotFoundError:: Captura el error que se produce si el archivo de entrada no se encuentra.









```python
import os  # Importamos el módulo os para trabajar con rutas de archivos y directorios.

def convertir_a_mayusculas():
    # Definimos la ruta completa del archivo de entrada.
    # Es importante utilizar una ruta válida según el sistema operativo.
    archivo_entrada = r"C:\Users\sala2.akademos\Documents\lenguajes de programacion\CRUD Raigosa Miguel\codigo\archivo_entrada.txt"
    
    # Definimos la ruta del archivo de salida. 
    # Utilizamos os.path.join para crear una ruta que sea compatible con el sistema operativo.
    archivo_salida = os.path.join(os.path.dirname(archivo_entrada), 'salida.txt')
    
    try:
        # Abrimos el archivo de entrada en modo lectura.
        # 'encoding' especifica la codificación de caracteres (utf-8 es común para textos).
        with open(archivo_entrada, 'r', encoding='utf-8') as infile:
            # Leemos todo el contenido del archivo y lo almacenamos en la variable 'contenido'.
            contenido = infile.read()
        
        # Convertimos el contenido a mayúsculas utilizando el método upper().
        contenido_mayusculas = contenido.upper()
        
        # Abrimos el archivo de salida en modo escritura.
        # Esto creará el archivo 'salida.txt' en el mismo directorio que el archivo de entrada.
        with open(archivo_salida, 'w', encoding='utf-8') as outfile:
            # Escribimos el contenido convertido a mayúsculas en el archivo de salida.
            outfile.write(contenido_mayusculas)

        # Imprimimos un mensaje confirmando que el proceso ha sido exitoso.
        print(f"El contenido de '{archivo_entrada}' ha sido escrito en '{archivo_salida}' en mayúsculas.")

    except FileNotFoundError:
        # Manejo de excepciones: si el archivo de entrada no se encuentra, se muestra un mensaje.
        print(f"El archivo '{archivo_entrada}' no se encuentra.")
    except IsADirectoryError:
        # Manejo de excepciones: si la ruta es un directorio y no un archivo, se informa al usuario.
        print(f"Se esperaba un archivo, pero se encontró un directorio: '{archivo_entrada}'.")
    except Exception as e:
        # Capturamos cualquier otra excepción y mostramos el mensaje de error.
        print(f"Ocurrió un error: {e}")

# Llamamos a la función para ejecutar el proceso de conversión.
convertir_a_mayusculas()
```
